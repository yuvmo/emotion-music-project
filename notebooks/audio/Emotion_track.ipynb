```python
genre_list_df = pd.read_csv("genres_list.csv")
genre_list_df.head()
```

```python
def auto_genre_mapping(genre: str):
    g = genre.lower()

    # SAD
    if ("sad" in g) or ("dark" in g) or ("emo" in g) or ("melancholy" in g):
        return "sad"

    # HAPPY
    if ("pop" in g) or ("dance" in g) or ("edm" in g) or ("disco" in g) or ("house" in g):
        return "happy"

    # ANGRY
    if ("trap" in g) or ("drill" in g) or ("hard" in g) or ("metal" in g) or ("punk" in g):
        return "angry"

    # CALM
    if ("chill" in g) or ("lofi" in g) or ("lo-fi" in g) or ("ambient" in g) or ("piano" in g) or ("acoustic" in g):
        return "calm"

    # ROMANTIC
    if ("r&b" in g) or ("soul" in g) or ("ballad" in g) or ("indie r&b" in g):
        return "romantic"

    # NEUTRAL
    if ("rock" in g) or ("indie" in g) or ("folk" in g) or ("alternative" in g):
        return "neutral"

    return "unknown"
```

```python
genre_list_df["auto_emotion"] = genre_list_df["genre"].apply(auto_genre_mapping)
genre_list_df.head(20)
```

```python
def emotion_from_audio_features(valence, energy):
    if valence >= 0.6 and energy >= 0.6:
        return "happy"
    if valence <= 0.4 and energy <= 0.4:
        return "sad"
    if valence >= 0.6 and energy <= 0.4:
        return "calm"
    if valence <= 0.4 and energy >= 0.6:
        return "angry"
    return "neutral"
```

```python
import ast

genre_to_emotion = dict(zip(genre_list_df["genre"], genre_list_df["auto_emotion"]))

def assign_final_emotion(row):
    genres_str = row["genres"]
    valence = row["valence"]
    energy = row["energy"]

    try:
        genres = ast.literal_eval(genres_str)
        genre_emotions = [genre_to_emotion.get(g, "unknown") for g in genres]

        for e in genre_emotions:
            if e != "unknown":
                return e
    except:
        pass

    return emotion_from_audio_features(valence, energy)

tracks["emotion_final"] = tracks.apply(assign_final_emotion, axis=1)
```
