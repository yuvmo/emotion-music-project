```python
from lingua import Language, LanguageDetectorBuilder

languages = [
    Language.ENGLISH,
    Language.RUSSIAN,
    Language.UKRAINIAN,
    Language.GERMAN,
    Language.FRENCH,
    Language.SPANISH,
    Language.ITALIAN,
    Language.TURKISH,
    Language.CHINESE,
    Language.JAPANESE,
    Language.KOREAN,
]

detector = LanguageDetectorBuilder.from_languages(*languages).build()
```

```python
def detect_lang(text):
    if not isinstance(text, str) or len(text.strip()) == 0:
        return "unknown"

    lang = detector.detect_language_of(text)
    return lang.iso_code_639_1.name.lower() if lang else "unknown"
```

```python
tracks["lang"] = tracks["text"].apply(detect_lang)
```

```python
def map_group(lang):
    if lang == "ru":
        return "russian"
    if lang == "en":
        return "english"
    return "other"

tracks["lang_group"] = tracks["lang"].apply(map_group)
```

```python
artists_set = set()

for row in tracks["artists"]:
    try:
        for a in ast.literal_eval(row):
            artists_set.add(a)
    except:
        pass

len(artists_set)
```

```python
artist_lang = {}
```

```python
for artist in artists_set:
    if re.search("[А-Яа-яЁё]", artist):
        artist_lang[artist] = "russian"
```

```python
from lingua import Language, LanguageDetectorBuilder

detector = LanguageDetectorBuilder.from_all_languages().build()

def detect_artist_lang(artist):
    lang = detector.detect_language_of(artist)
    if not lang:
        return None
    return lang.iso_code_639_1.name.lower()
```

```python
for artist in artists_set:
    if artist in artist_lang:
        continue
    lang = detect_artist_lang(artist)
    artist_lang[artist] = lang
```

```python
rus_hint = [
    "miyagi", "andypanda", "hammali", "navai", "zivert", "rauf", "faik",
    "pharaoh", "kizaru", "markul", "big baby tape", "og buda", "lizer",
    "eldzhey", "guf", "timati", "mot", "feduk", "noize", "monetochka",
    "bi-2", "bi2", "gorky", "korzh", "korol", "shaman"
]

for artist in artists_set:
    low = artist.lower()
    if any(key in low for key in rus_hint):
        artist_lang[artist] = "russian"
```

```python
for artist, lang in artist_lang.items():
    if lang == "en":
        artist_lang[artist] = "english"
```

```python
for artist, lang in artist_lang.items():
    if lang is None or lang not in ["russian", "english"]:
        artist_lang[artist] = "other"
```

```python
def get_track_language(artists_str):
    artists = ast.literal_eval(artists_str)
    langs = [artist_lang[a] for a in artists if a in artist_lang]

    if "russian" in langs:
        return "russian"
    if "english" in langs:
        return "english"
    return "other"
```

```python
tracks["lang_artist_based"] = tracks["artists"].apply(get_track_language)
```
